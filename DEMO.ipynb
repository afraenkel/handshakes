# Project OFFHAND: Predicting gestures with Android Watch
* Watches reduce friction, allowing for seamless user-engagement. 
* Watches are more likely to belong to the identity using the watch than a phone/computer.

## Product Ideas:
* Fast, personal gesture-based passwords.  Paying with your phone? Have the usual password input replaced with a gesture when in range of the watch.
* Easy gesture-based user input.  Report an accident on waze with a flick of the wrist, instead of texting and driving.

## Current App:
We recorded 4 different gestures:
* Handshake, 
* Fistbump, 
* Slap-shake, 
* Hi Five

and used accelerometer data from an Android watch to recognize those gestures.

* A very rough version is already installed on our Android Watch. . .

# DEMO: The program illustrated in a python notebook:

```python
% matplotlib inline
from hndshk import *
import os
```

## Load Accelerometer Watch Data

```python
train_filelist = [ x for x in os.listdir('./') if ('.txt' in x and '__2' in x) ]
filelist = [ x for x in os.listdir('./') if ('.txt' in x and '__' in x) ]
```

```python
trainingD = load_gestures(train_filelist,preprocess=True)
testingD = load_gestures(filelist,preprocess=False)
```

## Sample Gestures -- what's a gesture's signature?

```python
plot_gestures(trainingD)
```

## Read & process a long stream of gestures

```python
streamD = create_streaming_gesture(testingD)  # create a stream of gestures
L,ans = streamD['data'],streamD['ans']
convDict = convolve_stream(L,trainingD)             # convolve with 'model' gestures
```

```python
L.plot(figsize=(16,4),title='Stream of Gestures')
```

```python
L.ix[5500:7500].plot(figsize=(16,4),title='close up on a few gestures')
```
